```python
# 비율/수치형 컬럼들 결측값 처리

#결측 플래그 생성
df_eda_cut['revol_util_f'] = df_eda_cut['revol_util'].isna().astype(int)
df_eda_cut['pct_tl_nvr_dlq_f'] = df_eda_cut['pct_tl_nvr_dlq'].isna().astype(int)
df_eda_cut['tot_cur_bal_f'] = df_eda_cut['tot_cur_bal'].isna().astype(int)
df_eda_cut['num_accts_ever_120_f'] = df_eda_cut['num_accts_ever_120_pd'].isna().astype(int)

# revol_util의 경우, revol_val > 0인 경우엔 중앙값으로 대체, revoㅣ_val = 0일 때 0인 값으로 대체
df_eda_cut.loc[(df_eda_cut['revol_bal']>0),'revol_util'] = df_eda_cut.loc[(df_eda_cut['revol_bal']>0),'revol_util'].fillna(df_eda_cut['revol_util'].median())
df_eda_cut.loc[(df_eda_cut['revol_bal']==0),'revol_util'] = df_eda_cut.loc[(df_eda_cut['revol_bal']==0),'revol_util'].fillna(0)

# 중앙값으로 대체
df_eda_cut['pct_tl_nvr_dlq'] = df_eda_cut['pct_tl_nvr_dlq'].fillna(df_eda_cut['pct_tl_nvr_dlq'].median())
df_eda_cut['tot_cur_bal'] = df_eda_cut['tot_cur_bal'].fillna(df_eda_cut['tot_cur_bal'].median())
df_eda_cut['num_accts_ever_120_pd'] = df_eda_cut['num_accts_ever_120_pd'].fillna(df_eda_cut['num_accts_ever_120_pd'].median())

# 데이터가 1개, 신용등급과 점수, 다른 컬럼들을 통하여 0으로 채워도 무방하겠다 판단. 
df_eda_cut['inq_last_6mths'] = df_eda_cut['inq_last_6mths'].fillna(0)
```

- annual_inc (연간소득) 컬럼 이상치 처리

연간소득 최대값 10,999,200 (한화 약 160억원)
0.995 = 상위 0.5% 경계값보다 큰 값은 모두 상위 0.5%의 경계값으로 clipping

```python
# 상위 0.5% 클립
upper = df_eda_cut['annual_inc'].quantile(0.995)
df_eda_cut['annual_inc_clip'] = df_eda_cut['annual_inc'].clip(upper=upper)

# 로그 변환으로 분포 완화
df_eda_cut['annual_inc_log'] = np.log1p(df_eda_cut['annual_inc_clip']) 
```

- dti 컬럼 이상치 처리

```python
# 값 : -1 1개나옴
# 연소득으로 계산한 월소득과 월 대출상환만 
# 계산해도 6%이상으로 나와서 0 중앙값으로 처리

df_eda_cut['dti'] = df_eda_cut['dti'].replace(-1, np.nan)
df_eda_cut['dti'] = df_eda_cut['dti'].fillna(df_eda_cut['dti'].median())
```

```python
# home_ownership 표본 통일
rare = ["ANY","NONE"]

df_eda_cut["home_ownership"] = df_eda_cut["home_ownership"].replace(
    rare,
    "OTHER"
)
```

```python
# 결측 및 이상치 처리 저장
df_clean = df_eda_cut.copy()
```

-------------------------------------------------------------
### 파생컬럼 생성

```python
# 파생컬럼 - 계좌당 잔액 (빚)
df_clean['avg_bal_per_acc'] = df_clean['tot_cur_bal'] / df_clean['total_acc'].replace(0, np.nan)
```

```python
# 구간으로 된 신용점수 평균값으로 나타내기
df_clean['fico_avg'] = (df_clean['fico_range_low'] + df_clean['fico_range_high']) / 2
```

```python
# [3/5 전처리] fico 그룹핑 + 파생컬럼
df_clean['fico_group'] = pd.cut(
df_clean['fico_avg'],
bins=[300, 580, 670, 740, 800, 850],
labels=['poor','fair','good','very good','excellent'])
```

```python
# 첫 신용 계좌 생성부터의 기간

issue = df_clean['issue_d']
earliest = df_clean['earliest_cr_line']

# 파생컬럼 값이 정수로 나오도록 연 월 구분하여 계산하여 기간을 개월수에 대한 정수로 계산
df_clean['cr_hist_months'] = (issue.dt.year - earliest.dt.year) * 12 + (issue.dt.month - earliest.dt.month)
```

```python
 # 파생컬럼 - 계좌당 잔액 (빚) [사용]
df['avg_bal_per_acc'] = df['tot_cur_bal'] / df['total_acc'].replace(0, np.nan)
#전체 신용계좌중에서 한번이라도 120이상 연체된 계좌의 비율 [미사용]
df['pd120_ratio'] = df['num_accts_ever_120_pd'] / df['total_acc'].replace(0, np.nan) 
```

```python
df['loan_status'].value_counts()
```

```python
# 소득 대비 대출금 비율 파생컬럼
df_clean["loan_income_ratio"] = df_clean["loan_amnt"] / df_clean["annual_inc_clip"]
```

```python
# emp_length 그룹 파생컬럼
def group_emp_length(x):
    if x in ['< 1 year', '1 year', '2 years']: return '0-2년(저연차)'
    elif x in ['3 years', '4 years', '5 years']: return '3-5년(중간)'
    elif x in ['6 years', '7 years', '8 years', '9 years']: return '6-9년(중상)'
    elif x == '10+ years': return '10년+(고연차)'
    else : return 'Unknown'

df_clean['emp_length_group'] = df_clean['emp_length'].apply(group_emp_length)
```

```python
#대출상태 분류 
# Fully Paid → 0
# Charged Off / Default → 1

df_loan_status = df_clean.loc[df_clean["loan_status"].isin(["Fully Paid", "Charged Off", "Default"])].copy()
df_current = df_clean.loc[df_clean["loan_status"] == 'Current'].copy()
df_loan_status["default_rate"] = (df_loan_status["loan_status"] != "Fully Paid").astype(int)
```
